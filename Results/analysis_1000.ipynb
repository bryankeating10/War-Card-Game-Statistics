# **Analyzing 1000 game simulation results**

```python
import pandas as pd
df_1000 = pd.read_csv('test_1000.csv')
df_1000.set_index('game_num',inplace=True)

display(df_1000.head())
```

```python
# Distibution of rounds per game
N = 5000 # Number of moves constituting an infinite game
rounds_exc_inf = df_1000[df_1000['rounds'] < N] # Rounds excluding infinite games
not_infinite = (df_1000['rounds'] < N).sum() # Number of games under N limit

print(f'{not_infinite} games under {N} round limit')
rounds_exc_inf['rounds'].plot.hist(bins=200,title='Rounds per game')
```
